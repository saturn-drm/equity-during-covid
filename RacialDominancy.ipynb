```python
race =  pd.read_csv('datas/6-RaceandHispanicOrigin.csv' ,usecols=[2,9,10,11,12,13,14,15,16,17], skiprows=0)
race = race.groupby(['modzcta']).sum()
race['Percent of Population that is Non-Hispanic White'] = race['Non-Hispanic White Population']/race['Total Population']*100
race['Percent of Population that is Black or African American'] = race['Black or African American Population']/race['Total Population']*100
race['Percent of Population that is American Indian and Alaska Native'] = race['American Indian and Alaska Native Population']/race['Total Population']*100
race['Percent of Population that is Asian'] = race['Asian Population']/race['Total Population']*100
race['Percent of Population that is Native Hawaiian and Other Pacific Islander'] = race['Native Hawaiian and Other Pacific Islander Population']/race['Total Population']*100
race['Percent of Population that is Some Other Race'] = race['Some Other Race Population']/race['Total Population']*100
race['Percent of Population that is 2 or More Races'] = race['Two or More Races Population']/race['Total Population']*100
race['Percent of Population that is Hispanic or Latino'] = race['Hispanic or Latino Population']/race['Total Population']*100
race=race.drop(['Total Population',
                'Non-Hispanic White Population',
                'Black or African American Population',
                'American Indian and Alaska Native Population',
                'Asian Population',
                'Native Hawaiian and Other Pacific Islander Population',
                'Some Other Race Population',
                'Two or More Races Population',
                'Hispanic or Latino Population'], axis=1)
race['maxperc'] = race.max(axis=1)
race['maxrace'] = race.idxmax(axis=1)
race.to_csv('datas/6-RaceandHispanicOriginMerged.csv', columns=['maxperc','maxrace'],index=True)
```
