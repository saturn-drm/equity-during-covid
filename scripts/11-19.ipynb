```python
colTOdrop = []
for column in incomeraw.columns:
    if 'Margin of Error' in column:
        colTOdrop.append(column)
incomeColDropped = incomeraw.drop(columns=colTOdrop, inplace=False)
incomeColDropped.head()
```

```python
cleanedconvertor = convertor[['ZCTA5', 'GEOID']]
```

```python
testGEOID = cleanedconvertor[cleanedconvertor['ZCTA5'] == '10119']['GEOID'].iloc[0]
```

```python
incomeNY = incomeColDropped[incomeColDropped['State'] == 'New York']
incomeNY.shape
```

```python
incomeNY[incomeNY['Geographic Identifier - FIPS Code'] == testGEOID]
```

```python
incomeNY['ZCTA5'] = ''
incomeNY['moczcta'] = ''
incomeNY.shape
```

```python
incomeNYC = pandas.DataFrame(columns=incomeNY.columns)
for zcta in zcta_modzcta:
    geoidlist = cleanedconvertor[cleanedconvertor['ZCTA5'] == zcta]['GEOID'].to_list()
    for geoid in geoidlist:
        tmpseires = incomeNY[incomeNY['Geographic Identifier - FIPS Code'] == geoid]
        tmpseires['ZCTA5'] = zcta
        tmpseires['moczcta'] = zcta_modzcta[zcta]
        incomeNYC = incomeNYC.append(tmpseires, ignore_index=True)
incomeNYC
```

```python
incomeNYC.reset_index(drop=True, inplace=True)
incomeNYC.head()
```

```python
colToMove1 = incomeNYC.pop('ZCTA5')
colToMove2 = incomeNYC.pop('modzcta')
incomeNYC.insert(1, 'ZCTA5', colToMove1)
incomeNYC.insert(2, 'modzcta', colToMove2)
incomeNYC.head()
```

```python
incomeNYC.to_csv('../datas/01income-cleaned.csv', index=False)
```
